```python
import pandas as pd
videos = pd.read_csv("shortform_videos.csv")
creators = pd.read_csv("shortform_creators.csv")
platforms = pd.read_csv("shortform_platforms.csv")

videos.head()
```

```python
# Calculate retention rate and sort top-performing videos
videos['retention_rate'] = videos['full_views'] / videos['views']
videos[['video_id', 'retention_rate']].sort_values(by='retention_rate', ascending=False).head()
```

```python
# Analyze average performance by format type
videos.groupby('format_type')[['views', 'likes', 'comments', 'shares', 'retention_rate']].mean().sort_values(by='retention_rate', ascending=False)
```

```python
# Merge video and creator data
merged = videos.merge(creators, on='creator_id')
creator_summary = merged.groupby('creator_name')[['views', 'likes', 'shares', 'retention_rate']].mean()
creator_summary.sort_values(by='retention_rate', ascending=False)
```

```python
# Evaluate hook performance across videos
videos[['video_id', 'hook_watch_rate']].sort_values(by='hook_watch_rate', ascending=False).head()
```
